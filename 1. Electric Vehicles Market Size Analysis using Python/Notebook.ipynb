The dataset is from: https://statso.io/market-size-of-evs-case-study/

The provided dataset contains the following columns, each representing different aspects of the electric vehicle (EV) population in the United States:

- **VIN (1-10)**: Partial Vehicle Identification Number.
- **County**: The county in which the vehicle is registered.
- **City**: The city in which the vehicle is registered.
- **State**: The state in which the vehicle is registered. It appears that this dataset may be focused on Washington (WA) state.
- **Postal Code**: The postal code where the vehicle is registered.
- **Model Year**: The year of the vehicle model.
- **Make**: The manufacturer of the vehicle.
- **Model**: The model of the vehicle.
- **Electric Vehicle Type**: The type of electric vehicle, e.g., Battery Electric Vehicle (BEV).
- **Clean Alternative Fuel Vehicle (CAFV) Eligibility**: Eligibility status for clean alternative fuel vehicle programs.
- **Electric Range**: The maximum range of the vehicle on a single charge (in miles).
- **Base MSRP**: The Manufacturer’s Suggested Retail Price.
- **Legislative District**: The legislative district where the vehicle is registered.
- **DOL Vehicle ID**: Department of Licensing Vehicle Identification.
- **Vehicle Location**: Geographic coordinates of the vehicle location.
- **Electric Utility**: The electric utility service provider for the vehicle’s location.
- **2020 Census Tract**: The census tract for the vehicle’s location.

## EV Adoption Over Time
Analyze the growth of the EV population by model year.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

# EV Adoption Over Time
plt.figure(figsize = (12, 6))
ev_adoption_by_year = ev_data['Model Year'].value_counts().sort_values()
print(ev_adoption_by_year.head())

sns.barplot(x = ev_adoption_by_year.index, y = ev_adoption_by_year.values, palette = 'viridis')
plt.title('EV Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above bar chart:
1. Early Growth Phase (1997 - 2015):
* From 1997 to around 2015, EV registrations show a modest growth trend. The adoption rate during this period is relatively low, indicating the early stages of the EV market.
* This phase likely corresponds to the initial market entry of EVs, where consumers and manufacturers were beginning to explore electric mobility options.
2.	Acceleration Phase (2016 - 2021):
* Starting around 2016, there is a noticeable upward trend in EV registrations. This period marks a significant acceleration in the adoption of EVs.
* The increase in registrations during these years can be attributed to several factors, including advancements in EV technology, improved battery life, increased availability of models, and growing consumer awareness of environmental benefits.
3.	Exponential Growth Phase (2022 - 2023):
* The chart shows a dramatic spike in EV registrations in 2023, with the number of registered vehicles reaching a peak. This exponential growth highlights a critical point of market adoption.
* This surge may be driven by stronger government incentives, more comprehensive charging infrastructure, significant improvements in vehicle range, and a broader acceptance of EVs as a mainstream transportation option.

5. **Low Frequency of High Ranges**: Despite the existence of electric vehicles with ranges extending up to around 350 miles, these high-range vehicles are much less frequent. The density curve smoothens these observations, showing that they are outliers compared to the overall distribution.

6. **Distribution Insights**: The high frequency of low-range vehicles combined with the low mean range implies that the dataset might include a significant number of older models, short-range electric vehicles, or hybrid vehicles with limited electric-only capability. The presence of vehicles with ranges up to 350 miles indicates advancements in battery technology, but these models are not yet prevalent.

### Forecasting Future EV Registrations in 2024

We'll calculate the Compound Annual Growth Rate (CAGR) between a recent year with complete data (2023) and an earlier year to project the 2024 figures. Additionally, using this growth rate, we can estimate the market size for the next five years.

```python
from scipy.optimize import curve_fit
import numpy as np

# filter the dataset to include years with complete data
filtered_years = ev_registration_counts[ev_registration_counts.index <= 2023]

# Define a function for exponential growth to fit the data
def exponential_growth(x, a, b):
    return a * np.exp(b * x)

# prepare the data for curve fitting
x_data = filtered_years.index - filtered_years.index.min()
y_data = filtered_years.values

display(x_data, y_data)
print('-'*100, '\n', 'Then we can fit the data to the exponential growth function using the curve_fit() function from the scipy.optimize module.')

# fit the data to the exponential growth function
params, covariance = curve_fit(exponential_growth, x_data, y_data)
# Simply to display the results
print(curve_fit(exponential_growth, x_data, y_data))

# Use the fitted function to forecase the number of EV registrations for 2024 and the next 5 years
forecast_years = np.arange(2024, 2024 + 6) - filtered_years.index.min() # np.arange(start, stop, step)
forecast_values = exponential_growth(forecast_years, *params) # *params - unpack the parameters from the tuple

# create a dictionary to store the forecasted values
forecasted_EVs = dict(zip(np.arange(2024, 2024 + 6), forecast_values))
display(forecasted_EVs)
```

### Explanation of the forecast_values part

```python
# fit the data to the exponential growth function
params, covariance = curve_fit(exponential_growth, x_data, y_data)
display(params, covariance) # Note that covariance is never used in the code

# Use the fitted function to forecase the number of EV registrations for 2024 and the next 5 years
forecast_years = np.arange(2024, 2024 + 6) - filtered_years.index.min() # np.arange(start, stop, step)

def exponential_growth(x, a, b):
    return a * np.exp(b * x)
forecast_values = exponential_growth(forecast_years, *params) # *params - unpack the parameters from the tuple
display(forecast_values)

# Alternatively, we can directly calculate the forecast values using the parameters
# a, b = *params = 1.1008124, 0.41415396
forecast_values = 1.1008124 * np.exp(0.41415396 * np.array([27, 28, 29, 30, 31, 32]))
display(forecast_values)
```

### Plot the Estimated Market Size data

```python
years = np.arange(filtered_years.index.min(), 2024 + 6)
actual_years = filtered_years.index
forecast_years_full = np.arange(2024, 2024 + 6)
display(years, actual_years, forecast_years_full)

# Actual and forecasted values
actual_values = filtered_years.values
forecasted_values_full = [forecasted_EVs[year] for year in forecast_years_full] # where forecasted_EVs is the dictionary containing the forecasted values
display(actual_values, forecasted_values_full)

# Plotting the actual and forecasted values
plt.figure(figsize=(12, 8))
plt.plot(actual_years, actual_values, 'bo-', label='Actual Registrations')
plt.plot(forecast_years_full, forecasted_values_full, 'ro--', label='Forecasted Registrations')

plt.title('Current & Estimated EV Market')
plt.xlabel('Year')
plt.ylabel('Number of EV Registrations')
plt.legend()
plt.grid(True)

plt.show()


```

From the above chart, we observe:
* The number of actual EV registrations remained dorman until around 2010, after which a consistent and steep increase in registrations is observed.
* The forecasted EV registrations predict a more dramtic increase in the coming years.
